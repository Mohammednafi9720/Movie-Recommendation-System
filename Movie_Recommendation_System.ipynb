```python
import pandas as pd

movies = {
    'Movie': ['Inception', 'Titanic', 'Avengers', 'Interstellar', 'Joker'],
    'Genre': ['Sci-Fi', 'Romance', 'Action', 'Sci-Fi', 'Drama'],
    'Rating': [9, 8, 9, 9, 8]
}

df = pd.DataFrame(movies)

print("Movie Dataset")
print(df)

recommended = df[df['Rating'] >= 9]

print("\nRecommended Movies:")
print(recommended[['Movie', 'Genre']])
```
